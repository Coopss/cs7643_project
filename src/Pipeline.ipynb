```python
from transformers import pipeline

chef = pipeline('text-generation', model='./model', tokenizer='gpt2', config={'max_length':1500})

chef("How was your day? <BOS>")
```

```python
model = BertForSequenceClassification.from_pretrained('upvote_model', num_labels=4)

tokenizer = BertTokenizerFast.from_pretrained('bert-base-cased', use_cache=True)
```

```python
encodings = tokenizer(text=["How was your day? <BOS>"], text_pair=['Not to bad, just the first time I saw reddit after going to sleep last night.\nIt was just me playing games and browsing the subreddit on my own.\nI watched the whole series last night'], truncation=True, padding=True)
```

```python
model.forward(encodings)
```
